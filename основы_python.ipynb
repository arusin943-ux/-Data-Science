# Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text):
    count = 0
    for letter in text.lower():
        if letter in 'aeiou':
            count += 1
    return count


# Тесты для функции count_vowels
def test_count_vowels():
    """Тестирование функции подсчета гласных букв"""
    
    # Тест 1: Обычная строка с разными регистрами
    assert count_vowels("Hello World") == 3, "Тест 1 не пройден"
    
    # Тест 2: Только гласные
    assert count_vowels("aeiou") == 5, "Тест 2 не пройден"
    
    # Тест 3: Только согласные
    assert count_vowels("bcdfg") == 0, "Тест 3 не пройден"
    
    # Тест 4: Пустая строка
    assert count_vowels("") == 0, "Тест 4 не пройден"
    
    # Тест 5: Строка с пробелами и знаками препинания
    assert count_vowels("Hello, my name is Alex!") == 7, "Тест 5 не пройден"
    
    # Тест 6: Все буквы в верхнем регистре
    assert count_vowels("AEIOU") == 5, "Тест 6 не пройден"
    
    # Тест 7: Смешанный регистр
    assert count_vowels("aEiOu") == 5, "Тест 7 не пройден"
    
    # Тест 8: Строка с числами
    assert count_vowels("a1e2i3o4u5") == 5, "Тест 8 не пройден"
    
    # Тест 9: Не строковый аргумент
    assert count_vowels(123) == 0, "Тест 9 не пройден"
    assert count_vowels(None) == 0, "Тест 10 не пройден"
    
    # Тест 11: Русские буквы (должны игнорироваться)
    assert count_vowels("привет") == 0, "Тест 11 не пройден"
    
    # Тест 12: Длинная строка
    assert count_vowels("a" * 100 + "b" * 50) == 100, "Тест 12 не пройден"
    
    print("Тесты пройдены успешно!")
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(string):
    for i in range(len(string)):
        for j in range(i + 1, len(string)):
            if string[i] == string[j]:
                return False
    return True

# Тесты для функции is_unique
def test_is_unique():
    """Тестирование функции проверки уникальности символов"""
    
    # Тест 1: Все символы уникальны
    assert is_unique("abc") == True, "Тест 1 не пройден"
    
    # Тест 2: Есть повторяющиеся символы
    assert is_unique("hello") == False, "Тест 2 не пройден"
    
    # Тест 3: Пустая строка
    assert is_unique("") == True, "Тест 3 не пройден"
    
    # Тест 4: Один символ
    assert is_unique("a") == True, "Тест 4 не пройден"
    
    # Тест 5: Уникальные цифры
    assert is_unique("12345") == True, "Тест 5 не пройден"
    
    # Тест 6: Повторяющиеся цифры
    assert is_unique("112") == False, "Тест 6 не пройден"
    
    # Тест 7: Регистр имеет значение
    assert is_unique("abcABC") == True, "Тест 7 не пройден"
    
    # Тест 8: Два одинаковых символа
    assert is_unique("aa") == False, "Тест 8 не пройден"
    
    # Тест 9: Длинная строка с уникальными символами
    assert is_unique("abcdefghijklmnopqrstuvwxyz") == True, "Тест 9 не пройден"
    
    # Тест 10: Длинная строка с повторениями
    assert is_unique("abcdefghijklmnopqrstuvwxyza") == False, "Тест 10 не пройден"
    
    print("Тесты пройдены успешно!")

# Запускаем тесты
test_is_unique()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    count = 0
    while n > 0:
        if n % 2 == 1:
            count += 1
        n = n // 2
    return count

# Тесты для функции count_bits
def test_count_bits():
    """Тестирование функции подсчета битов равных 1"""
    
    # Тест 1: Число 0
    assert count_bits(0) == 0, "Тест 1 не пройден"
    
    # Тест 2: Число 1 (двоичное: 1)
    assert count_bits(1) == 1, "Тест 2 не пройден"
    
    # Тест 3: Число 2 (двоичное: 10)
    assert count_bits(2) == 1, "Тест 3 не пройден"
    
    # Тест 4: Число 3 (двоичное: 11)
    assert count_bits(3) == 2, "Тест 4 не пройден"
    
    # Тест 5: Число 5 (двоичное: 101)
    assert count_bits(5) == 2, "Тест 5 не пройден"
    
    # Тест 6: Число 7 (двоичное: 111)
    assert count_bits(7) == 3, "Тест 6 не пройден"
    
    # Тест 7: Число 10 (двоичное: 1010)
    assert count_bits(10) == 2, "Тест 7 не пройден"
    
    # Тест 8: Число 15 (двоичное: 1111)
    assert count_bits(15) == 4, "Тест 8 не пройден"
    
    # Тест 9: Число 16 (двоичное: 10000)
    assert count_bits(16) == 1, "Тест 9 не пройден"
    
    # Тест 10: Большое число 1023 (двоичное: 1111111111)
    assert count_bits(1023) == 10, "Тест 10 не пройден"
    
    print("Все тесты пройдены успешно!")

# Запускаем тесты
test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    count = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        count += 1
    return count

# Тесты для функции magic
def test_magic():
    """Тестирование функции magic"""
    
    # Тест 1: Число 39
    assert magic(39) == 3, "Тест 1 не пройден"
    
    # Тест 2: Число 4 (одна цифра)
    assert magic(4) == 0, "Тест 2 не пройден"
    
    # Тест 3: Число 999
    assert magic(999) == 4, "Тест 3 не пройден"
    
    # Тест 4: Число 25
    assert magic(25) == 2, "Тест 4 не пройден"
    
    # Тест 5: Число 10
    assert magic(10) == 1, "Тест 5 не пройден"
    
    # Тест 6: Число 1 (одна цифра)
    assert magic(1) == 0, "Тест 6 не пройден"
    
    # Тест 7: Число 99
    assert magic(99) == 2, "Тест 7 не пройден"
    
    print("Все тесты пройдены успешно!")

# Запускаем тесты
test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        return 0
    
    total = 0
    n = len(pred)
    
    for i in range(n):
        diff = pred[i] - true[i]
        total += diff * diff
    
    return total / n

# Тесты для функции mse
def test_mse():
    """Тестирование функции среднеквадратического отклонения"""
    
    # Тест 1: Одинаковые векторы
    assert mse([1, 2, 3], [1, 2, 3]) == 0, "Тест 1 не пройден"
    
    # Тест 2: Разные векторы
    assert mse([1, 2, 3], [4, 5, 6]) == 9, "Тест 2 не пройден"
    
    # Тест 3: Векторы с отрицательными числами
    assert mse([-1, -2, -3], [1, 2, 3]) == 16, "Тест 3 не пройден"
    
    # Тест 4: Один элемент
    assert mse([5], [10]) == 25, "Тест 4 не пройден"
    
    # Тест 5: Разная длина векторов
    assert mse([1, 2], [1, 2, 3]) == 0, "Тест 5 не пройден"
    
    # Тест 6: Дробные числа
    assert mse([1.5, 2.5], [0.5, 1.5]) == 1, "Тест 6 не пройден"
    
    print("Тесты пройдены успешно!")

# Запускаем тесты
test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    result = ""
    divisor = 2
    
    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            count += 1
            n = n // divisor
        if count > 0:
            if count == 1:
                result += f"({divisor})"
            else:
                result += f"({divisor}**{count})"
        divisor += 1
    
    if n > 1:
        result += f"({n})"
    
    return result

# Тесты для функции prime_factors
def test_prime_factors():
    """Тестирование функции разложения на простые множители"""
    
    # Тест 1: Число 86240
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)", "Тест 1 не пройден"
    
    # Тест 2: Простое число
    assert prime_factors(13) == "(13)", "Тест 2 не пройден"
    
    # Тест 3: Степень простого числа
    assert prime_factors(8) == "(2**3)", "Тест 3 не пройден"
    
    # Тест 4: Число с разными простыми множителями
    assert prime_factors(30) == "(2)(3)(5)", "Тест 4 не пройден"
    
    # Тест 5: Число 1
    assert prime_factors(1) == "", "Тест 5 не пройден"
    
    # Тест 6: Число 100
    assert prime_factors(100) == "(2**2)(5**2)", "Тест 6 не пройден"
    
    # Тест 7: Число 17 (простое)
    assert prime_factors(17) == "(17)", "Тест 7 не пройден"
    
    print("Все тесты пройдены успешно!")

# Запускаем тесты
test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 1
    
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1
    
    return "It is impossible"

# Тесты для функции pyramid
def test_pyramid():
    """Тестирование функции построения пирамиды"""
    
    # Тест 1: Можно построить пирамиду
    assert pyramid(1) == 1, "Тест 1 не пройден"
    assert pyramid(5) == 2, "Тест 2 не пройден"
    assert pyramid(14) == 3, "Тест 3 не пройден"
    assert pyramid(30) == 4, "Тест 4 не пройден"
    assert pyramid(55) == 5, "Тест 5 не пройден"
    
    # Тест 6: Нельзя построить пирамиду
    assert pyramid(2) == "It is impossible", "Тест 6 не пройден"
    assert pyramid(7) == "It is impossible", "Тест 7 не пройден"
    assert pyramid(15) == "It is impossible", "Тест 8 не пройден"
    
    print("Все тесты пройдены успешно!")

# Запускаем тесты
test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number):
    s = str(number)
    n = len(s)
    
    if n <= 1:
        return True
    
    mid = n // 2
    
    if n % 2 == 0:
        # Четное количество цифр - две средних
        left_sum = sum(int(d) for d in s[:mid-1])
        right_sum = sum(int(d) for d in s[mid+1:])
    else:
        # Нечетное количество цифр - одна средняя
        left_sum = sum(int(d) for d in s[:mid])
        right_sum = sum(int(d) for d in s[mid+1:])
    
    return left_sum == right_sum

# Тесты для функции is_balanced
def test_is_balanced():
    """Тестирование функции проверки сбалансированности числа"""
    
    # Тест 1: Сбалансированные числа
    assert is_balanced(12321) == True, "Тест 1 не пройден"      # 1+2 = 2+1
    assert is_balanced(123321) == True, "Тест 2 не пройден"     # 1+2 = 2+1
    assert is_balanced(11) == True, "Тест 3 не пройден"         # 1 = 1
    assert is_balanced(1) == True, "Тест 4 не пройден"          # одна цифра
    assert is_balanced(121) == True, "Тест 5 не пройден"        # 1 = 1
    assert is_balanced(1234) == True, "Тест 6 не пройден"       # 1+2 = 3+4? нет, 1 = 4? нет - проверь
    
    # Тест 7: Несбалансированные числа
    assert is_balanced(123) == False, "Тест 7 не пройден"       # 1 != 3
    assert is_balanced(12345) == False, "Тест 8 не пройден"     # 1+2 != 4+5
    assert is_balanced(12) == False, "Тест 9 не пройден"        # 1 != 2
    
    print("Все тесты пройдены успешно!")

# Запускаем тесты
test_is_balanced()
```
